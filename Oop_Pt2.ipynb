```python
class Animal:
  def speak(self):
    print("Animal Speaking")

class Dog(Animal):
  def speak(self):
    print("Dog speaking")

class Cat(Animal):
  def speak(self):
    print("Cat speaking")

```

```python
obj1 = Dog()
obj2 = Cat()
```

```python
obj1.speak()
```

```python
class Animal:
  def speak(self, sound, animal):
    print(f"The {animal} says {sound}")

  def speak1(self, animal,activity):
    print(f"The {animal} is doing {activity}")
```

```python
obj.speak("Meow", "Cat")
```

```python
obj.speak1("Dog", "Walking")
```

```python
class Animal:
  a = 10 #Gloabal variable
  def __init__(self, sound, animal, activity):
    self.sound = sound #self. value contains the vlaue of that parmeter
    self.animal = animal
    self.activity = activity


  def speak(self):#NO need to give any parameters
  #Foloows reusablity
    print(f"The {self.animal} says {self.sound}")

  def speak1(self):
    print(f"The {self.animal} is doing {self.activity}")

obj = Animal("Woof", "Dog", "Walking")
obj.speak1()
```

```python
obj = Animal()
```

```python
obj = Animal("Woof", "Dog", "Walking")
```

```python
obj.speak1()
```

```python
class Electronics:
  def __init__(self, mobile, laptop):
    self.mobile = mobile
    self.laptop = laptop

    print(f"The mobile is {self.mobile} and laptop is {self.laptop}")

  def __init__(self, speaker, microphone):
    self.speaker = speaker
    self.microphone = microphone

    print(f"The speaker is {self.speaker} and microphone is {self.microphone}")
```

```python
obj = Electronics("Samsung", "Dell") #OVERRIDING
```
