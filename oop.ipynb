```python
var = 100  # Global variable

def fun1():
    var = 20  # Local variable in fun1
    print(f"Value before fun2 call: {var}")

    def fun2():
        global var  # Refers to the global 'var' outside of all functions
        var = 30  # Updates the global variable

    fun2()  # Call fun2
    print(f"Value after fun2 call in fun1: {var}")  # 'var' in fun1 remains unchanged

print(f"Value before fun1 call: {var}")  # Global variable
fun1()
print(f"Value after fun1 call: {var}")  # Global variable updated by fun2
```

# Inheritence

```python
class Father(): #Base/Parent Class

  def write(self):
    return "Yes"

class Son(Father): #derived class #child class
  def write(self):
    return "Yes"
```

```python
#Thereare three types of inheritence
#1. Single Inheritence
#2. Multiple Inheritence
#3. Multilevel Inheritence all can use in python but only 1 & 3 are allowed in java
```

# Single Inheritence

```python
#Single inheritence says that one class is inherting from other
```

```python
class Animal:
  def speak(self):
    print("Animal is speaking")

class Dog(Animal):
  def bark(self):
    print("Dog is barking")
  #def speak(self):
   # print("Animal is speaking")
  def speak(self):
    print("Cat is speaking")
```

```python
obj = Dog()
```

```python
obj.bark()
```

#Multiple Inheritence

```python
'''
#class 1  --->   class 2 --->     class 3 ---> Class 4
------------------------------------------------^

'''
```

```python
class Samp1:
  def new1(self):
    print( " First class ")
  def new2(self):
    print("First class")

class Sampl2:

  def new1(self):
    print("Second class")
  def new2(self):
    print("second class")

class Finale(Samp1, Sampl2):
  def eat(self):
    print("Eating")
  def new1(self):
    print( " child class ")
```

# Method Overriding

```python
class Animal:
  def speak(self):
    return 10

class Dog(Animal):
  def speak(self):
    return 7

class Cat(Animal):
  def speak(self):
    return 2
```

```python
obj1 = Animal()
obj2 = Dog()
obj3 = Cat()
```

```python
obj3.speak() #even if the name of the function is same still it will override
```
