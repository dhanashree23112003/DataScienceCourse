```python
#A Function -> age (input)
#  if age > 60 -> Old
#if the age is b/w 30 to 60 -> Middle aged
#if the age is b/w 19 to 30 -> Young
# if the age is b/w 13 to 19 -> Teen
#else  -> kid
```

```python
def Age(age_value):          #Fitrst define the function
  if age_value > 60:
    print("Old")

#  elif age_value >= 30 and age_value <= 60:
  elif 30 < age_value <= 60:
    #print("Middle aged")
     return "Middle aged"
  elif 19 < age_value <= 30:
    return "Young"
    #print("Young")

  elif 13 < age_value <= 19:
    return "Teen"
    #print("Teen")

  else:
    return "Kid"
    #print("Kid")
```
